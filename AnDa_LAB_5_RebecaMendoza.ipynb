## **Relaciones en distribuciones de datos**
Encontrar relaciones entre variables es un proceso fundamental en la ciencia de datos. Las técnicas estadísticas y de aprendizaje automático, como la correlación, la regresión y el análisis de redes, se utilizan para descubrir estas relaciones, a menudo visualizando los datos de diversas maneras.
A medida que la cantidad y complejidad de los datos siguen creciendo, la importancia de encontrar relaciones entre variables seguirá aumentando. La capacidad de identificar estas relaciones será crucial para resolver problemas complejos, tomar decisiones informadas y avanzar en el conocimiento en diversos campos.

Encontrar relaciones entre los datos nos puede ayudar en aspectos como:
- Explicar Fenómenos
- Hacer Predicciones
- Tomar Decisiones Informadas



## **Valores Faltantes**
Los valores faltantes son un problema común en los conjuntos de datos del mundo real. Pueden deberse a diversas causas, como errores de entrada de datos, respuestas no proporcionadas en encuestas o la naturaleza misma del fenómeno que se está midiendo. Si no se abordan adecuadamente, los valores faltantes pueden tener un impacto significativo en los análisis de datos, lo que lleva a resultados sesgados e inexactos.

**¿Por qué es importante el manejo de valores faltantes?**
Existen varias razones por las que el manejo de valores faltantes es crucial en la ciencia de datos:
1. **Reduce el sesgo:** Cuando se ignoran los valores faltantes, se asume implícitamente que esos casos son similares a los casos completos. Esta suposición puede ser errónea y conducir a sesgos en los resultados del análisis. Por ejemplo, si un conjunto de datos sobre los ingresos tiene muchos valores faltantes para los ingresos altos, la media del ingreso se subestimará si se ignoran estos valores faltantes.
2. **Aumenta la precisión:** La imputación de los valores faltantes puede aumentar la precisión de los análisis estadísticos al aumentar el tamaño de la muestra y reducir la variabilidad. Esto es particularmente importante para los conjuntos de datos pequeños o cuando se utilizan técnicas de análisis estadístico que son sensibles al tamaño de la muestra.
3. **Mejora la interpretabilidad:** Los modelos de aprendizaje automático pueden ser sensibles a la presencia de valores faltantes. La imputación de los valores faltantes puede ayudar a mejorar el rendimiento del modelo y facilitar la interpretación de los resultados.
4. **Permite un análisis más completo:** Al tratar los valores faltantes, se puede utilizar toda la información disponible en un conjunto de datos, lo que permite un análisis más completo y preciso.


## **¿Qué es la imputación?**
La imputación es un proceso estadístico para estimar los valores faltantes en un conjunto de datos. Existen diversos métodos de imputación, cada uno con sus propias ventajas e inconvenientes. La elección del método de imputación adecuado depende del patrón de falta de datos, la naturaleza de las variables y el objetivo del análisis.

Elegir el método de imputación adecuado, ni un método de imputación único que sea mejor para todas las situaciones. La elección del método adecuado depende de varios factores, como el patrón de falta de datos, la naturaleza de las variables y el objetivo del análisis. Es importante considerar cuidadosamente las opciones disponibles y seleccionar el método que sea más probable que produzca resultados precisos y no sesgados.


En resumen, el encontrar relaciones,  el manejo de valores faltantes y la imputación son aspectos cruciales de la ciencia de datos. La exploración de los datos nos permite entender a priori su comportamiento y contenido.  Al abordar adecuadamente los valores faltantes, se puede garantizar que los análisis de datos sean precisos, confiables y útiles para la toma de decisiones. La correcta imputación de valores asegura que el conjunto de datos siga manteniendo los mismos estadísticos y características originales sin afectar los análsis y relaciones.

## Funciones de probabilidad de masas (PMFs)

## **[5]** Hacer un pmf de bill_length_mm y obtener la probabilidad para 3 valores diferentes, presente la gráfica de la distribución

```python
from palmerpenguins import load_penguins

# Cargar el dataset de los pingüinos
penguins = load_penguins()

# Eliminar filas con valores faltantes en 'bill_length_mm'
penguins = penguins.dropna(subset=['bill_length_mm'])

# Crear una PMF para bill_length_mm usando sns.histplot
plt.figure(figsize=(10, 6))
sns.histplot(data=penguins, x='bill_length_mm', binwidth=1, stat='probability', color='blue', alpha=0.7)

# Título y etiquetas del gráfico
plt.title('Función de Masa de Probabilidad de bill_length_mm')
plt.xlabel('Longitud del Pico (mm)')
plt.ylabel('Probabilidad')

# Seleccionar tres valores diferentes de bill_length_mm para calcular sus probabilidades
valores = [40.1, 45.2, 50.3]
probabilidades = penguins['bill_length_mm'].value_counts(normalize=True).sort_index().reindex(valores, fill_value=0)

# Mostrar las probabilidades de los tres valores seleccionados
print("Probabilidades para los valores seleccionados:")
for valor, probabilidad in probabilidades.items():
    print(f"Valor: {valor} mm, Probabilidad: {probabilidad:.4f}")

# Añadir líneas y etiquetas para los valores seleccionados en el gráfico
for valor in valores:
    plt.axvline(x=valor, color='red', linestyle='--')
    plt.text(valor, plt.ylim()[1] * 0.9, f'{valor}\n{probabilidades[valor]:.4f}', 
             color='red', ha='center')

plt.show()
```

## Funciones empirícas de probabilidad acumulada (ECDFs)

## **[6]** Hacer un ecdf de bill_depth_mm y obtener la probabilidad para 3 valores diferentes. Muestre la Gráfica resultante

```python
from empiricaldist import Cdf
from palmerpenguins import load_penguins

# Cargar el conjunto de datos de los pingüinos
penguins = load_penguins()

# Calcular la ECDF para la variable bill_length_mm
ecdf_bill_length_mm = Cdf.from_seq(penguins['bill_length_mm'])

# Graficar la ECDF
plt.figure(figsize=(10, 6))
ecdf_bill_length_mm.plot()
plt.title('Función de Distribución Acumulativa Empírica de bill_length_mm')
plt.xlabel('Longitud del Pico (mm)')
plt.ylabel('Probabilidad Acumulada')

# Calcular la probabilidad para tres valores diferentes
valores = [40, 45, 50]
probabilidades = {}
for valor in valores:
    probabilidades[valor] = ecdf_bill_length_mm.forward(valor)
    plt.plot(valor, probabilidades[valor], 'ro', label=f'Probabilidad en {valor}')

plt.legend()
plt.show()
```

## **[36]** Desarrolle un proceso de imputación mediante KNN(Vecinos Cercanos) para las variables de *ozone* y *solar_r*. Muestre la gráfica donde indica los valores imputados, realice el proceso usando un k = 5 y k = 13
